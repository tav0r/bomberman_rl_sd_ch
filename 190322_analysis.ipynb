## check movement patterns
is the board explored by different kind of movements of does it get stuck somewhere

```python
action_options = ["LEFT", "UP", "RIGHT", "DOWN", "WAIT","BOMB"]
```

```python
actions = []
with open(r'C:\Users\chris\Documents\Code\IFML-WS2018-19\FinalProject\bomberman_rl-master\agent_code\neo\logs\neo.log', 'r') as log:
    allstr = ''
    for line in log:
        allstr += line
        if 'LEFT' in line:
            actions.append(0)
        elif 'UP' in line:
            actions.append(1)
        elif 'RIGHT' in line:
            actions.append(2)
        elif 'DOWN' in line:
            actions.append(3)
        elif 'WAIT' in line:
            actions.append(4)
        elif 'BOMB' in line:
            actions.append(5)
```

```python
plt.plot(actions)
plt.yticks(range(len(action_options)), action_options)
plt.show()
fig, ax = plt.subplots()
n, bins, patches = plt.hist(actions, bins=[0,1,2,3,4,5,6], linewidth=0, )
ax.set_xticks(bins+0.5)
ax.set_xticklabels(action_options)
plt.show()
```

```python
# look into one specific episode
n = 50
actions_episode = actions[n*400:(n+1)*400]

fig, ax = plt.subplots()
n, bins, patches = plt.hist(actions_episode, bins=[0,1,2,3,4,5,6], linewidth=0, )
ax.set_xticks(bins+0.5)
ax.set_xticklabels(action_options)
plt.show()
```
